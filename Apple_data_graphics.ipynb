```python
appl = pd.read_csv('datasets/AAPL.csv', header=0, index_col='date', parse_dates=True, names = col_names)
appl.head()
```

```python
import matplotlib.pyplot as plt
```

```python
close_series = appl['close']
close_series.plot()
plt.show()
```

```python
appl.plot()
plt.yscale('log')
plt.show()
```

```python
appl.loc['2009':'2010', ['high', 'low']].plot()
plt.yscale('log')
plt.show()
```

```python
appl['open'].plot(color='b', style='.-', legend=True)
appl['close'].plot(color='r', style='.', legend=True)
plt.axis(('1989','1991',0,2.5))
plt.show()
```
